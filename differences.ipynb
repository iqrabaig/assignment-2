```python
import numpy as np
import matplotlib.pyplot as plt
f= lambda x:0.1*x**5-0.2*x**3+0.1*x-0.2
h=0.1
x=np.linspace(-1,1,50)   # will create array of element between -1 and 1 having 50 equal 
# central difference

dff1= (f(x+h)-f(x-h))/h
dff2= (f(x+h)-2*f(x)+f(x-h))/h**2

#plot
plt.plot(x,f(x),'-k',x,dff1,'--b',x,dff2,'-.r')
plt.xlabel('x')
plt.xlabel('y')
plt.legend(["f(x)","f'(x)","f''(x)'"])
plt.grid()
```
