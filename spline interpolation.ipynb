# linear interpolation method 1

```python
import numpy as np 
import matplotlib.pyplot as plt
x= np.linspace(0,2**np.pi,10)         #10 equidisatnt points      
y=np.cos(x)                           # original x,y value

plt.plot(x,y,"o:")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()

x1=np.linspace(0,2**np.pi,100)           # values of x1 for which we will find y1 by interpolation
y1=np.interp(x1,x,y)  # given discrete data point x and y return a linear interpolate value at x1
plt.plot(x1,y1,"-x")           #plot interpolated points
plt.show()
plt.plot(x,y,"o:",x1,y1,"-x")  # plot both graphs in 1
plt.legend(["oringinal values","interpolated values"])    # shows what o represent and what -x represent in graph
```

```python
import numpy as np 
import matplotlib.pyplot as plt
x= np.linspace(0,2**np.pi,10)         #10 equidisatnt points      
y=np.cos(x)                           # original x,y value


x1=np.linspace(0,2**np.pi,100)           # values of x1 for which we will find y1 by interpolation
y1=np.interp(x1,x,y)  # given discrete data point x and y return a linear interpolate value at x1
y3=np.interp(2,x,y)    # interpolated value at 2
print("x=2 , y =  %f" % (y3))
y3=np.interp(10,x,y)    # interpolated value at 10
print("x=10 , y =  %f" % (y3))
xinput= float(input("enter the value for interpolation:"))    # taking value from user for interpolation
y4=np.interp(xinput,x,y)   
print("x= %d , y =  %f" % (xinput ,y4))
```

```python
from scipy.interpolate import interp1d
x=np.linspace(0,20,10)
y=x**3
x1=np.linspace(0,20,100)
f=interp1d(x,y,"linear")     # we can also find cubic, quardatic ,nearest,zero interpolation by just writing required interpolation in place of linear but default is linear
y1=f(x1)
plt.plot(x1,y1,"o:")
plt.grid()
plt.show()
x2=float(input("enter the value for interpolation:"))
y2=f(x2)
print("x= %f , y= %d" %(x2,y2))
```
